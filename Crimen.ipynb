##  1.- Los 5 tipos de crimenes que se presentaron con mas frencuencia en el mes de Julio para el estado de Aguascalientes entre el 2015 y el 2018 


```python
aguascalientes_anio = mexico[(mexico['year'] >= 2015) & (mexico['year']<=2018)]
```

```python
aguascalientes_anio
```

```python
aguascalientes_estado = aguascalientes_anio[(aguascalientes_anio['entity'] == 'Aguascalientes')]
```

```python
aguascalientes_estado
```

```python
aguascalientes_mes = aguascalientes_estado[(aguascalientes_estado['month'] == 'July')]
```

```python
aguascalientes_mes.columns = aguascalientes_mes.columns.str.replace('count','Conteo')
```

```python
aguascalientes_mes
```

```python
aguascalientes_crimen = aguascalientes_mes.groupby('type_of_crime')['Conteo'].sum().reset_index()
```

```python
aguascalientes_crimen.rename(columns = {'Conteo':'Total_Crimenes'}, inplace = True)
aguascalientes_crimen.rename(columns = {'type_of_crime':'Tipo_de_Crimenes'}, inplace = True)
```

```python
aguascalientes_crimen
```

```python
top_5_crimenes = aguascalientes_crimen.nlargest(5,'Total_Crimenes')
```

```python
top_5_crimenes.set_index('Tipo_de_Crimenes',inplace = True)
```

```python
top_5_crimenes.rename(index = {
    'Robbery' : 'Robo',
    'Injuries' : 'Lesiones',
    'Property damage' : 'Daños a la propiedad',
    'Other common law crimes' : 'Otros delitos comunes',
    'Threats' : 'Amenazas'
}, inplace = True )
```

```python
top_5_crimenes
```

```python
grafico_crimenes = top_5_crimenes.plot(
    kind = 'pie',
    y = 'Total_Crimenes',
    startangle = 90,
    autopct = '%1.1f%%',
    figsize = (15,13)
)

grafico_crimenes.set_title(
    'Mayor cantidad de crimenes : 2015 a 2018',
    fontdict = {
        'size' : 15,
        'weight' : 'bold',
        'color':'black'
    }
)


grafico_crimenes.set_ylabel(
    'Estado Aguascalientes - Julio',
    fontdict = {
        'size' : 12,
        'color' : 'black'
    }
)


grafico_crimenes.legend(
    labels = ['Robo: 4664','Lesiones: 1381','Daños a la propiedad: 1049','Otros Delitos Comunes: 813','Amenazas : 669'],
    fontsize = 8,
    labelcolor = ['black','black']
)
```

## 2.-Los 5 tipos de criemens que se presentaron con menor frecuenia entre el 2015 y el 2022 en la republica mexicana

```python
periodo_tiempo= mexico[(mexico['year'] >= 2015) & (mexico['year']<=2022)]
```

```python
periodo_tiempo.columns = mexico.columns.str.replace('type_of_crime','Tipo_de_Crimen')
```

```python
periodo_tiempo
```

```python
crimenes_republica = periodo_tiempo.groupby('Tipo_de_Crimen')['Conteo'].sum().reset_index()
```

```python
crimenes_republica
```

```python
crimen_menor_frecuencia = crimenes_republica.sort_values(by='Conteo')
```

```python
crimen_menor_frecuencia.rename(index = {
    'Incest' : 'Incesto',
    'Child trafficking' : 'Trafico de menores',
    'Prisoner escape' : 'Escape de prisioneros',
    'Abduction' : 'Abducción',
    'Human trafficking' : 'Trafico humano'
}, inplace = True )
```

```python
menores_crimenes = crimen_menor_frecuencia.nsmallest(5,'Conteo')
```

```python
#### OBSERVAMOS EL CRIMEN QUE HA SIDO MENOS PRESENTE EN LA REPUBLICA MEXICANA
```

```python
menores_crimenes
```

```python
grafico_menor_frecuencia = menores_crimenes.plot(
    kind = 'pie',
    y = 'Conteo',
    startangle = 90,
    autopct = '%1.1f%%',
    figsize = (15,13)
)


grafico_menor_frecuencia.set_title(
    'Crimenes menos frecuentes México: 2015-2022',
    fontdict = {
        'size' : 15,
        'weight' : 'bold',
        'color':'black'
    }
)


grafico_menor_frecuencia.set_ylabel(
    'Conteo de Crimenes',
    fontdict = {
        'size' : 12,
        'color' : 'black'
    }
)


grafico_menor_frecuencia.legend(
    labels = ['Incesto: 176','Trafico de niños: 607','Escape de prisioneros: 942','Abducción: 1317','Trafico humano : 4027'],
    fontsize = 8,
    labelcolor = ['black','black']
)
```

## 3.- ¿Qué tan comunes fueron los crimenes electorales en los estados de la republica mexicana en el año 2018?, teniendo en cuenta que en ese año fueron las elecciones presidenciales

```python
crimenes_electorales_anio = mexico[(mexico['year'] == 2018) ]
```

```python
crimenes_electorales_anio
```

```python

Crimenes_electorales = crimenes_electorales_anio[(crimenes_electorales_anio['type_of_crime'] == 'Electoral crimes') ]
```

```python
Crimenes_electorales
```

```python
crimenes_estado = Crimenes_electorales.groupby('entity')['Conteo'].sum().reset_index()
```

```python
crimenes_estado.set_index('entity',inplace = True)
```

```python
crimenes_estado.rename(index = { 'Ciudad de México' : 'CDMX',
                                 'Coahuila de Zaragoza' : 'Coahuila',
                                 'Michoacán de Ocampo' : 'Michoacán',
                                 'Veracruz de Ignacio de la Llave':'Veracruz',
                                 'México' : 'Edo. México'
                               },inplace = True)
```

```python
crimenes_estado
```

```python
estados_renombrados = crimenes_estado
```

```python
conteo_crimen = estados_renombrados.sort_values(by='Conteo')
```

```python
conteo_crimen
```

```python
grafico_electoral = conteo_crimen.plot(
    kind='bar',
    y='Conteo',
    color = '#fe5151', 
    figsize =(11,5)
)



plt.style.use(
    'seaborn-v0_8-pastel'
)


grafico_electoral.set_title(
    'Crimenes Electorales - 2018',
    fontdict = {
       'size' : 5,
        'weight' : 'bold',
        'color' : 'black'
    }
)




grafico_electoral.set_ylabel(
    'Numero de Crimenes',
    fontdict ={
        'size' : 12,
        'color' : 'black'
    }
)


grafico_electoral.set_xlabel(
    'Estado',
    fontdict ={
        'size' : 12,
        'color' : 'black'
    }
)


grafico_electoral.legend(
    labels = ['Conteo de Crimenes Electorales'],
    fontsize = 10,
    labelcolor = ['black','black']
)

grafico_electoral.annotate(
    '708',
    xy = (30,710.00),
    color = 'black'
)


grafico_electoral.annotate(
    '302',
    xy = (29,309.00),
    color = 'black'
)

grafico_electoral.annotate(
    '118',
    xy = (26,122.00),
    color = 'black'
)

plt.yticks(
    color = 'black'
)

plt.xticks(
    color = 'black'
)

grafico_electoral.xaxis.grid(
    color = 'black',
    linewidth = 0.3,
    linestyle = 'dashdot'
)

grafico_electoral.yaxis.grid(
    color = 'black',
    linewidth =  0.3,
    linestyle = 'dashdot'
)
```

## 4.- 6 Entidades Federativas con mas homicidios

```python
entidades = mexico[ (mexico['type_of_crime'] == 'Homicide') & 
            (mexico['entity'].isin(['Guanajuato','México','Baja California Sur','Chihuahua','Michoacán de Ocampo','Jalisco']))
]
```

```python
entidades
```

```python
crecimiento = entidades.groupby(['year','entity']).agg({'Conteo': 'sum'})
```

```python
crecimiento
```

```python
reacomodo_columnas = crecimiento.reset_index(1).pivot(columns = 'entity').droplevel(0,axis = 'columns')
```

```python
reacomodo_columnas
```

```python
plt.style.use(
    'seaborn-v0_8-poster'
)

entidades_mayores_homicidios = reacomodo_columnas.plot(
    kind = 'line',
    figsize = (12,10)
)

entidades_mayores_homicidios.set_title(
    'Crecimiento de Homicidios: 2015-2022',
    fontdict = {
       'size' : 9,
        'weight' : 'bold',
        'color' : 'black'
    }
)


entidades_mayores_homicidios.set_ylabel(
    'Numero de Crimenes',
    fontdict ={
        'size' : 12,
        'color' : 'black'
    }
)


entidades_mayores_homicidios.set_xlabel(
    'Años',
    fontdict ={
        'size' : 12,
        'color' : 'black'
    }
)

entidades_mayores_homicidios.legend(
    labels = ['B.C Sur','Chihuahua','Guanajuato','Jalisco','Michoacán','México'],
    fontsize = 10,
    labelcolor = ['Black','Black']
)


entidades_mayores_homicidios.xaxis.grid(
    color = 'black',
    linewidth = 0.3,
    linestyle = 'dashdot'
)

entidades_mayores_homicidios.yaxis.grid(
    color = 'black',
    linewidth =  0.3,
    linestyle = 'dashdot'
)

#B.C Sur
entidades_mayores_homicidios.annotate(
    '693',
    xy = (2017,800.00),
    color = 'black'
)

#Chihuahua
entidades_mayores_homicidios.annotate(
    '2388',
    xy = (2020,2400),
    color = 'black'
)

#Guanajuato
entidades_mayores_homicidios.annotate(
    '4940',
    xy = (2020,5000.00),
    color = 'black'
)


#Jalisco
entidades_mayores_homicidios.annotate(
    '2762',
    xy = (2018,2800),
    color = 'black'
)


#Michoacan
entidades_mayores_homicidios.annotate(
    '3237',
    xy = (2021,3300),
    color = 'black'
)

#Edo Mex
entidades_mayores_homicidios.annotate(
    '3237',
    xy = (2022,3900),
    color = 'black'
)


plt.yticks(
    color = 'black'
)

plt.xticks(
    color = 'black'
)
```

Tras analizar esta grafico observamos que los homicidios han ido bajando en el año 2023, lo cual contrasta con la informacion obtenida en la pagina que se encuentra al inicio de este documento, el conjunto de datos no es el ideal para analizar esta problematica, debido a que si existiera una baja en los homicidios de estas entidades, dichas entidades no estarian sumergidas en la violencia.
